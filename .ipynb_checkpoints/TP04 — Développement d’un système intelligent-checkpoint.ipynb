```python
categories = ('BIG_BAND', 'BLUES_CONTEMPORARY', 'COUNTRY_TRADITIONAL', 'DANCE', 'ELECTRONICA', 'EXPERIMENTAL',
              'FOLK_INTERNATIONAL', 'GOSPEL', 'GRUNGE_EMO', 'HIP_HOP_RAP', 'JAZZ_CLASSIC', 'METAL_ALTERNATIVE',
              'METAL_DEATH', 'METAL_HEAVY', 'POP_CONTEMPORARY', 'POP_INDIE', 'POP_LATIN', 'PUNK', 'REGGAE',
              'RNB_SOUL', 'ROCK_ALTERNATIVE', 'ROCK_COLLEGE', 'ROCK_CONTEMPORARY', 'ROCK_HARD', 'ROCK_NEO_PSYCHEDELIA')
```

```python
def split_function(nameof_csv):
    derivative_labels = pd.read_csv('tagged_feature_sets/'+nameof_csv+'/'+nameof_csv+'.csv',
                                    delimiter=',', header=None).values[:, -1:]
    for i, label in enumerate(derivative_labels) :
        derivative_labels[i] = categories.index(label)
    derivative_features = pd.read_csv('tagged_feature_sets/'+nameof_csv+'/'+nameof_csv+'.csv',
                                      delimiter=',', header=None).values[:, 2:-1]
    np.asarray(derivative_labels)
    np.asarray(derivative_features)
    derivative_labels = to_categorical(derivative_labels, num_classes=25)
    X_train, X_test, Y_train, Y_test = train_test_split(derivative_features, derivative_labels,
                                                                    test_size=0.20, random_state=42,
                                                                    stratify=derivative_labels)
    
    return X_train, X_test, Y_train, Y_test
```

# 3. Définition des différentes Modèles: Bayes Naif 

## 3.1 Gaussian

```python
def Bayes_model(X_train, X_test, Y_train, Y_test):
    accuracyp=[]
    F1_Scorep=[]
    X_train=X_train.astype(float)
    X_test=X_test.astype(float)
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.fit_transform(X_test)
    Y_train = argmax(Y_train, axis=1)
    Y_test = argmax(Y_test, axis=1)
    model = GaussianNB()
    modeli = GaussianNB()

    model = model.fit(X_train, Y_train)
    
    # Use the model to predict the class of samples
    # Notice that we are testing the train dataset
    Y_test_pred = model.predict(X_test)
    global acc_digits_data
    acc_digits_data = accuracy_score(Y_test, Y_test_pred )*100
    accuracyp.append(acc_digits_data)
    global f1_scor
#     print("Correct classification rate for the test dataset "+nameof_csv +"="+str(acc_digits_data*100)+"%")
    f1_scor=f1_score(Y_test, Y_test_pred, average="weighted")*100
    F1_Scorep.append(f1_scor)
    print("f1_score with Gaussian ="+str(f1_scor)+"%")

    print("Accuracy rate for the test dataset  = "+str(acc_digits_data))
    return modeli
```

### Précision et f1-score (Leave-one-out cross-validation)

```python
accbayes=[]
f1_scorebay=[]
for i in range(len(datacsv)):
    X_train, X_test, Y_train, Y_test=split_function(datacsv[i])
    Bayes_model( X_train, X_test, Y_train, Y_test)
    accbayes.append(acc_digits_data)
    f1_scorebay.append(f1_scor)
```

## 3.2 Multinomial :MinMaxScalar

```python
 #Discretisation

def multinomial_MinMaxScaler(X_train, X_test, Y_train, Y_test):
    discretizer = MinMaxScaler()
    Xp_D=discretizer.fit_transform(X_train)
    Y_train = argmax(Y_train, axis=1)
    Y_test = argmax(Y_test, axis=1)
    Xp_test_D=discretizer.fit_transform(X_test)
    model_multinomial = MultinomialNB(fit_prior=True)
    model_i = MultinomialNB(fit_prior=True)

    multinomial = model_multinomial.fit(Xp_D, Y_train)

    Yp_test_pred_D = multinomial.predict(Xp_test_D)
    global acc_digits_data
    acc_digits_data = accuracy_score(Y_test, Yp_test_pred_D )*100

    print("Correct classification rate for the test dataset = "+str(acc_digits_data)+"%")
    global f1_scor
    f1_scor=f1_score(Y_test, Yp_test_pred_D, average="weighted")*100
    print("f1_score with multinomial MinMaxScaler  = "+str(f1_scor)+"%")
    return model_i
```

### Précision et f1-score  (Leave-one-out cross-validation)

```python
accminmax=[]
f1_scoreminmax=[]
for i in range(len(datacsv)):
    X_train, X_test, Y_train, Y_test=split_function(datacsv[i])
    multinomial_MinMaxScaler( X_train, X_test, Y_train, Y_test)
    accminmax.append(acc_digits_data)
    f1_scoreminmax.append(f1_scor)
```

## 3.3 Multinomial: KBinsDiscretizer

```python

# transform the dataset with KBinsDiscretizer
def multinomial_KBinsDiscretizer(X_train, X_test, Y_train, Y_test):
    enc = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
    enc.fit(X_train)
    enc.fit(X_test)
    Y_train = argmax(Y_train, axis=1)
    Y_test = argmax(Y_test, axis=1)
    Xp_Dis = enc.transform(X_train)
    Xp_test_Dis = enc.transform(X_test)
    model_multinomial_K = MultinomialNB(fit_prior=True)
    multinomial_K = model_multinomial_K.fit(Xp_Dis, Y_train)
    Yp_test_pred_Dis = multinomial_K.predict(Xp_test_Dis)
    global acc_digits_data
    acc_digits_data = accuracy_score(Y_test, Yp_test_pred_Dis )

    print("Correct classification rate for the test dataset = "+str(acc_digits_data*100)+"%")
    global f1_scor
    f1_scor=f1_score(Y_test, Yp_test_pred_Dis, average="weighted")*100
    print("f1_score with MultinomialNB KBinsDiscretizer= "+str(f1_scor)+"%")
```

### Précision et f1-score  (Leave-one-out cross-validation)

```python
accKBins=[]
f1_scoreKBins=[]
for i in range(len(datacsv)):
    X_train, X_test, Y_train, Y_test=split_function(datacsv[i])
    multinomial_KBinsDiscretizer( X_train, X_test, Y_train, Y_test)
    accKBins.append(acc_digits_data)
    f1_scoreKBins.append(f1_scor)
```

# 4. Définition des différentes Modèles: Réseau de neurones

```python
def mdp_model():
    model = Sequential()
    model.add(Dense(100, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.2))

    model.add(Dense(75, activation='relu'))
    model.add(BatchNormalization())
    model.add(Dropout(0.2))

#     model.add(Dense(50, activation='relu'))
#     model.add(BatchNormalization())
#     model.add(Dropout(0.2))
    
#     model.add(Dense(30, activation='relu'))
#     model.add(BatchNormalization())
#     model.add(Dropout(0.2))

    model.add(Dense(25, activation='softmax'))
    return model
```

```python
def RNA_model(X_train, X_test, Y_train, Y_test):    
    X_train=X_train.astype(float)
    X_test=X_test.astype(float)
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.fit_transform(X_test)
    
    F1_score_test=[]
    acc_test=[]
    acc=[]
    val_acc=[]
    model =mdp_model()
    modeli =mdp_model()
    model.compile(optimizer=Adam(lr=5e-4, decay=5e-4 / 50),
                      loss='categorical_crossentropy',
                      metrics=['accuracy'])
    
    
    history = model.fit(X_train, Y_train, epochs=50, batch_size=500, validation_data=(X_test, Y_test), verbose = 0)
    y_probs = model.predict(X_test)
    y_pred = y_probs.argmax(axis=1)
    Y_testp=Y_test.argmax(axis=1)
    global k
    k=accuracy_score(Y_testp, y_pred)*100
    global f1_scor
    f1_scor=f1_score(Y_testp, y_pred, average="weighted")*100
    print("f1_score = "+str(f1_scor)+"%")
    print("Accuracy rate for the test dataset = "+str(k))
    plt.plot(history.history['acc'])  
    plt.plot(history.history['val_acc'])
    plt.title('model accuracy ')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'test'], loc='upper left')
    return modeli
```

# 5. Définition des différentes Modèles: K-plus proche voisin

```python

def KNN_model(X_train, X_test, Y_train, Y_test,n_neighbors):
    accuracyp=[]
    F1_Scorep=[]
    weights = 'uniform'
    metric = 'euclidean'
    algorithm = 'brute'

    model = neighbors.KNeighborsClassifier(n_neighbors, weights=weights, algorithm=algorithm, metric=metric )
    modeli = neighbors.KNeighborsClassifier(n_neighbors, weights=weights, algorithm=algorithm, metric=metric )

    model = model.fit(X_train, Y_train)
    # Use the model to predict the class of samples
    # Notice that we are testing the train dataset
    Y_test_pred = model.predict(X_test)
    global acc_digits_data
    acc_digits_data = accuracy_score(Y_test, Y_test_pred )
    accuracyp.append(acc_digits_data*100)
#     global f1_score
#     print("Correct classification rate for the test dataset "+nameof_csv +"="+str(acc_digits_data*100)+"%")
    global f1_scor
    f1_scor=f1_score(Y_test, Y_test_pred, average="weighted")
    F1_Scorep.append(f1_scor)
    print("f1_score with Gaussian ="+str(f1_scor)+"%")

    print("Accuracy rate for the test dataset  = "+str(acc_digits_data))
    return modeli
```

```python
model_RNA=model_RNA.compile(optimizer=Adam(lr=5e-4, decay=5e-4 / 50),
                      loss='categorical_crossentropy',
                      metrics=['accuracy'])
```

Combinaison entre les trois modèles

```python
# Building the Ensemble 
# Combination with MAJORITY VOTE (labels only)

def ENSEMBLE_model_3():
    print("Majority Vote of RNA, Bayes and KNN\n")
    model = VotingClassifier(estimators = [ ('RNA',model_RNA ), ('NB', model_Bayes) , ('KNN', model_KNN) ], 
                             voting = 'hard',
                             n_jobs = 3) 
    return model


def ENSEMBLE_model_2():
    print("Majority Vote of RNA and Bayes\n")
    model = VotingClassifier(estimators = [ ('RNA', model_RNA), ('KNN', model_KNN)], 
                             voting = 'hard',
                             n_jobs = 2) 
    return model
```

```python
# Build the ENSEMBLE model
modelENSEMBLE=ENSEMBLE_model_2()
datacsv = ['msd-jmirderivatives_dev', 'msd-jmirlpc_dev', 'msd-jmirmfccs_dev', 'msd-jmirmoments_dev', 'msd-jmirspectral_dev', 'msd-marsyas_dev_new', 'msd-mvd_dev', 'msd-rh_dev_new', 'msd-ssd_dev', 'msd-trh_dev']
for i in range(len(datacsv)):
    X_train, X_test, Y_train, Y_test=split_function(datacsv[i])
    # X_train=X_train.astype(float)
    # X_test=X_test.astype(float)
    # scaler = StandardScaler()
    # X_train = scaler.fit_transform(X_train)
    # X_test = scaler.fit_transform(X_test)
    Ytrain2 = argmax(Y_train, axis=1)
    Ytest2 = argmax(Y_test, axis=1)
    # Fit the Ensemble (TRAIN)
    modelENSEMBLE.fit(X_train, Ytrain2)
    # Use the model to predict the class of samples
    # Notice that we are testing on the 3 data splits
    # Use the ENSEMBBLE model to predict the class of samples
    # Notice that we are testing on the 3 data splits
    Ytrain_pred_ENSEMBLE = modelENSEMBLE.predict(X_train)
    Ytest_pred_ENSEMBLE  = modelENSEMBLE.predict(X_test)
    scores_tr_ENSEMBLE  = accuracy_score(Ytrain2, Ytrain_pred_ENSEMBLE )
    scores_tst_ENSEMBLE = accuracy_score(Ytest2,  Ytest_pred_ENSEMBLE )
    print("Classification rate for the training dataset (Ensemble)     = "+str(scores_tr_ENSEMBLE*100)+"%")
    print("Classification rate for the test dataset (Ensemble)         = "+str(scores_tst_ENSEMBLE*100)+"%")
```
